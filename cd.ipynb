```python
import pandas as pd
# Code for cleaning data from openrocket Cd as a mach function
# Initialize lists to store the data
power_on_rows = []
power_off_rows = []
all_rows = []

# Initial state: Engine is assumed ON (since data starts after Ignition)
engine_on = True 

input_filename = 'C:\\Users\\krikb\\Desktop\\simulations\\data\\cdmach.csv'

with open(input_filename, 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        
        # Check for comments and events
        if line.startswith('#'):
            # Detect the burnout event to switch to "Power Off" state
            if "BURNOUT" in line:
                engine_on = False
            continue
        
        # Parse data lines
        try:
            parts = line.split(',')
            if len(parts) >= 2:
                # User specified: Col 1 is Cd, Col 2 is Mach
                cd = float(parts[0])
                mach = float(parts[1])
                
                row = {'Mach': mach, 'Cd': cd}
                
                # Add to the global list
                all_rows.append(row)
                
                # Add to the specific phase list
                if engine_on:
                    power_on_rows.append(row)
                else:
                    power_off_rows.append(row)
        except ValueError:
            continue

# Create DataFrames
df_all = pd.DataFrame(all_rows)
df_on = pd.DataFrame(power_on_rows)
df_off = pd.DataFrame(power_off_rows)

# Save to CSV files
df_all.to_csv('cleaned_cd_mach.csv', index=False)
df_on.to_csv('powerondrag.csv', index=False)
df_off.to_csv('poweroffdrag.csv', index=False)

print("Files created successfully.")
```
