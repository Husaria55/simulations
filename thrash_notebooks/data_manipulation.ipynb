```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cog_df = pd.read_csv("center_of_pressure.csv")
x = cog_df["x"]
y = cog_df["Scalar"]
plt.plot(x, y)
plt.xlabel("Time (s)")
plt.ylabel("Center of Pressure Position (m)")
plt.title("Center of Pressure Position vs Time")
plt.grid()
plt.show()
```
